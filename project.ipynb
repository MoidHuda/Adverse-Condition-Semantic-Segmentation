```python
import os
```

```python
for cond in ["fog", "snow", "rain", "night"]:
    fog_dir_anon = f"data/acdc/rgb_anon/{cond}/"
    fog_dir_gt = f"data/acdc/gt/{cond}/"

    directory_anon = os.fsencode(fog_dir_anon)
    directory_gt = os.fsencode(fog_dir_gt)

    for split_a, split_g in zip([l for l in os.listdir(directory_anon) if l in [b'val', b'train']], os.listdir(directory_gt)):
            split_path_a = os.path.join(directory_anon, split_a)
            split_dir_a = os.fsencode(split_path_a)
            split_path_g = os.path.join(directory_gt, split_g)
            split_dir_g = os.fsencode(split_path_g)
            
            lst = open(f"data/list/acdc/{split_a.decode('utf-8')}.lst", "a")

            for city_a, city_g in zip(os.listdir(split_dir_a), os.listdir(split_dir_g)):
                city_path_a = os.path.join(split_path_a, city_a)
                city_dir_a = os.fsencode(city_path_a)
                city_path_g = os.path.join(split_path_g, city_g)
                city_dir_g = os.fsencode(city_path_g)

                for image_a, image_g in zip(os.listdir(city_dir_a), os.listdir(city_dir_g)):
                    if image_g.decode("utf-8").split("_")[-1] == "labelIds.png":
                        image_pt = os.path.join(city_path_a, image_a).decode('utf-8')
                        gt_pt = os.path.join(city_path_g, image_g).decode('utf-8')
                        lst.write( image_pt[5:] + "\t" + gt_pt[5:] + '\n')
                    
            lst.close()
        
```
